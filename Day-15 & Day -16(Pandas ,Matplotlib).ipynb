```python
import numpy as np
import matplotlib.pyplot as plt
x_axes=[1,2,3,4,5,6]
y_axes=[9,7,4,6,3,4]
plt.plot(x_axes,y_axes,'r')
plt.show()




```

```python
# pie chart:
import matplotlib.pyplot as plt
labels=['python','html','java','c','sql']
slices=[60000,29000,30000,40000,41000]
explode=[0,0,0.1,0,0]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},startangle=90,explode=explode,shadow=True)
plt.tight_layout()
plt.show()
```

```python
# pie chart:
import matplotlib.pyplot as plt
labels=['python','html','java','c','sql']
slices=[60000,29000,30000,40000,41000]
explode=[0,1,0,1,0]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':"yellow"},startangle=80,explode=explode,shadow=True)
plt.tight_layout()
plt.show()
```

```python
# basic linear plot
import matplotlib.pyplot as plt
x=[25,26,27,28,29]
y=[29000,30000,31000,32000,34000]
plt.plot(x,y,label='yaxis',color='#008000',linestyle='--',marker='o')
y1=[28000,32000,49000,50000,65000]
plt.plot(x,y1,label='y1axis',color='y',linestyle='-.',marker='d')
plt.tight_layout()
plt.title('basic linear plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.grid(True)
plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```
